```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_excel("amazon.xlsx")
```

```python
# KPI analysis
# Total sales
df['TotalAmount'].sum()
```

```python
# Total orders
df['OrderID'].count()
```

```python
# Average sales
df['TotalAmount'].mean()
```

```python
# Maximum quantity
df['Quantity'].max()
```

```python
# REVENUE BY STATE
df.groupby('State')['TotalAmount'].sum()
```

```python
# REVENUE BY CATEGORY
df.groupby('Category')['TotalAmount'].sum().sort_values(ascending=False)
```

```python
# TOP 5 CUSTOMERS
df.groupby('CustomerName')['TotalAmount'].sum().sort_values(ascending=False).head()
```

```python
# ORDERS BY PAYMENT MODE
df.groupby('PaymentMethod')['OrderID'].count()
```

```python
# Visulaization
df.groupby('State')['TotalAmount'].sum().plot(kind='bar')
plt.xlabel('State')
plt.ylabel('Revenue')
plt.show()
```

```python
df.groupby('State')['TotalAmount'].sum().sort_values(ascending=False).plot(kind='bar')
```

```python
df.groupby(['CustomerID','CustomerName'])['TotalAmount'].sum().sort_values(ascending=False).head(5)
```

```python
# Category wise sales
df.groupby('Category')['TotalAmount'].sum().sort_values(ascending=False).plot(kind='pie')
plt.title('Category wise sales')
plt.show()
```
